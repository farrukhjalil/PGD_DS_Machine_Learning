```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_Data.csv")
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
X=df[["YearsExperience"]].values
X
```

```python
Y_true=df[["Salary"]].values
Y_true
```

```python
print(X.shape)
print(Y_true.shape)
```

# Deep Learning Model

```python
model = Sequential()
model.add(Dense(1, input_shape=(1,)))
model.summary()
```

```python
model.compile(Adam(1800), 'mean_squared_error')
```

# Training

```python
model.fit(X, Y_true, epochs = 230)
```

```python
Y_pred = model.predict(X)
Y_pred
```

```python
Y_true
```

```python
df.plot(kind='scatter', x='YearsExperience', y='Salary', title ='Salary prediction based on provided experience')
plt.plot(X,Y_pred, color = 'red', linewidth=2)
```

# Trainable Parameters
